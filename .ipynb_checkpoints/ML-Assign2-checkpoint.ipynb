# InterQuartile(IQR) Method 

Since the Age column data is skewed we will apply IQR method here:

```python
# Question 8.
new_df['Age'].describe() #Here max age is 79 and min age is 0.
```

```python
sns.boxplot(new_df['Age'])

#Outliners are identified usig this boxplot.
```

```python
#Finding the IQR:

percentile25=new_df['Age'].quantile(0.25)
percentile75=new_df['Age'].quantile(0.75)
```

```python
percentile25
```

```python
percentile75
```

```python
iqr=percentile75-percentile25
```

```python
iqr
```

```python
upper_limit=percentile75 + 1.5*iqr
lower_limit=percentile25 - 1.5*iqr
```

```python
print("Upper_limit",upper_limit)
print("Lower limit",lower_limit)
```

```python
new_df.shape
```

**Finding Outliers-**

```python
new_df[new_df['Age']>upper_limit]
```

**Trimming the outliers-**

```python
new_df_cap=new_df.copy()

new_df_cap['Age']=np.where(new_df_cap['Age']>upper_limit, upper_limit,np.where(new_df_cap['Age']<lower_limit,lower_limit,new_df_cap['Age']))
```

```python
new_df_cap.describe()
```

```python
arr=new_df_cap['Age'].unique()
arr
```

```python
#Comparing
plt.figure(figsize=(16,8))
plt.subplot(2,2,1)
sns.distplot(new_df['Age'])

plt.subplot(2,2,2)
sns.boxplot(new_df['Age'])

plt.subplot(2,2,3)
sns.distplot(new_df_cap['Age'])

plt.subplot(2,2,4)
sns.boxplot(new_df_cap['Age'])

plt.show()
```
