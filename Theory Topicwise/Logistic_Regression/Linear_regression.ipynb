```python
import seaborn as sns
import pandas as pd
from sklearn import linear_model
```

```python
df=pd.read_csv(r'C:\Users\KIIT\MACHINE LEARNING\Insurance.csv')
```

```python
df
```

**Plotting the values on a graph using seaborn library:**

```python
sns.lmplot(x='Age',y='Premium',data=df)
```

```python
reg=linear_model.LinearRegression()
```

**Training the model with fit:**

```python
reg.fit(df[['Age']],df[['Premium']])
```

**Testing the model using prediction:**

```python
reg.predict([[23]])
```

```python
reg.predict([[50]])
```

**To find Coefficient/Gradient/Slope:**

```python
reg.coef_
```

**To find Intercept of line:**

```python
reg.intercept_
```
