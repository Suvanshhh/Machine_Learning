 2. ORDINAL ENCODING: Is used when there is a particular order to arrange the given data.
 3. LABEL ENCODING: Is same as Ordinal but is only used in case of Output Labels(y), while ordinal is only used in case of Input Labels(x).

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
# Here we are making the function of OneHotEncoder and dropping the two columns.
ohe = OneHotEncoder(drop='first',sparse=False,dtype=np.int32)
```

```python
# Fitting and Transforming Fuel and Owner column in the function.
X_train_new = ohe.fit_transform(X_train[['fuel','owner']])
```
